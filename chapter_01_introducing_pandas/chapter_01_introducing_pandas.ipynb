### 1.3.4	Filtering a Column by One or More Criteria

```python
movies[movies["Studio"] == "Universal"]
```

```python
released_by_universal = (movies["Studio"] == "Universal")
print(released_by_universal)
```

```python
## boolean indexing with bool series
movies[released_by_universal].head()
```

```python
# parallel ANDed boolean indexing
released_by_universal = movies["Studio"] == "Universal"
released_in_2015 = movies["Year"] == 2015
movies[released_by_universal & released_in_2015]
```

```python
# parallel ORed boolean indexing
released_by_universal = movies["Studio"] == "Universal"
released_in_2015 = movies["Year"] == 2015
movies[released_by_universal | released_in_2015]
```

```python
before_1975 = movies["Year"] < 1975
movies[before_1975]
```

```python
# can use .between for int dtype columns
# Other methods: https://pandas.pydata.org/docs/reference/series.html#computations-descriptive-stats
mid_80s = movies["Year"].between(1983, 1986)
movies[mid_80s]
```

```python
print(type(movies['Studio']))
print(type(movies['Studio'].str))
```

```python
has_dark_in_title = movies.index.str.lower().str.contains("dark")
has_para_in_studio = movies['Studio'].str.lower().str.contains('para')
movies[has_dark_in_title & has_para_in_studio]
```

### 1.3.5 Grouping Data

```python
movies["Gross"].str.replace(
    "$", "", regex = False
).str.replace(",", "", regex = False).astype(float).case_when(caselist=[
                                                  (lambda x: x>=3000, "HIT"),
                                                  (lambda x: (x<3000) & (x>=2000), "HIT"),
                                                  (lambda x: x<2000, "FLOP")
                                                  ]
                                                )
```

```python
(
    movies["Gross"]
    .str.replace("$", "", regex = False)
    .str.replace(",", "", regex = False)
    .astype(float)
)
```

```python
movies["Gross"] = (
    movies["Gross"]
    .str.replace("$", "", regex = False)
    .str.replace(",", "", regex = False)
    .astype(float)
)
```

```python
movies["Gross"].mean()
```

```python
studios = movies.groupby("Studio")
```

```python
studios.all().index
```

```python
studios["Gross"].sum().head()
```

```python
studios["Gross"].count().sort_values(ascending = False).head()
```

```python
studios["Gross"].sum().head()
```

```python
studios["Gross"].sum().sort_values(ascending = False).head()
```
